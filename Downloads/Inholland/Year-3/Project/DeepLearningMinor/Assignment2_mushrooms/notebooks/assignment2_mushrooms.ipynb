# Mushroom Classification using Deep Learning

## Objective
The goal of this project is to classify different types of mushrooms 
based on image data using deep learning techniques.

## Dataset Overview

The dataset consists of multiple folders, each representing a mushroom class:
- Agaricus
- Amanita
- Boletus
- Cortinarius
- Entoloma
- Hygrocybe
- Lactarius
- Russula
- Suillus

Each folder contains images of the respective mushroom type.
